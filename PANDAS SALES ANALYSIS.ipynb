<h4>EXPLORATORY DATA ANALYSIS

- <H5>WHAT IS THE OVERALL SALES TREND?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year from order_date
df['month-year'] = df['order_date'].apply(lambda x : x.strftime('%y-%m'))
```

```python
# Grouping month-year
df_trend = df.groupby('month-year').sum()['sales'].reset_index()
```

```python
# setting the figure size
plt.figure(figsize=(15,6))
plt.plot(df_trend['month-year'],df_trend['sales'],color='#A020F0')
plt.xticks(rotation='vertical',size=8,)
plt.show()
```

- <h5>WHICH ARE THE TOP 10 PRODUCTS BY SALES?

```python
# Grouping products by sales
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
# Sorting the dataframe in descending order
prod_sales = prod_sales.sort_values('sales',ascending=False)
```

```python
# Top 10 products by sales
prod_sales[:10]
```

- <h5> WHICH ARE THE MOST SELLING PRODUCTS?

```python
# Grouping products name
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
# Sorting the dataframe in descending order
most_sell_prod = most_sell_prod.sort_values('quantity',ascending = False)
```

```python
# Most selling products
most_sell_prod[:10]
```

- <h5> WHAT IS THE MOST PREFERRED SHIP MODE?

```python
# setting figure size
plt.figure(figsize=(8,6.5))

# Plotting shipmode
sns.countplot(df['ship_mode'])

plt.show()
```

- <h5> WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?

```python
# Grouping products by Category and Sub-Cate
cat_subcat_profit = pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
# Sorting the values
cat_subcat_profit.sort_values(['category','profit'],ascending=False)
```
